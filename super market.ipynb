# SuperMarket Data Analysis

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df =pd.read_csv("supermarket_sales - Sheet1.csv")
df.head()
```

```python
df =df.drop(columns = ["Invoice ID","Date","Time"],axis = 1)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df["gross margin percentage"].unique().sum()
```

```python
df = df.drop("gross margin percentage",axis= 1)
```

```python
df.head(5)
```

# Data Visualization

# Customer Gender 

```python
df["Gender"].value_counts()
```

```python
get_dummies = pd.get_dummies(df["Gender"])
get_dummies
```

```python
df = pd.concat([df , get_dummies ],axis = 1)
df.head()
```

```python
plt.figure(figsize = (12,6))

sns.barplot( x='Product line',  y='Female',data=df)
```

```python
plt.figure(figsize = (12,6))

sns.barplot( x='Product line',  y='Male',data=df)
```

# Customers City

```python
place = pd.DataFrame(df["City"].value_counts())
place
```

```python
sns.barplot(x=place.index , y= place["City"], palette = "hot")
plt.figure(figsize=(6,6))
```

# payment method

```python
payment = pd.DataFrame(df["Payment"].value_counts())
payment
```

```python
sns.barplot(x=payment.index, y=payment.Payment)
```

# Getting Gross Income For Each Product Line

```python
plt.figure(figsize=(12,6))
sns.barplot(x= df["Product line"], y = df["gross income"])
```
